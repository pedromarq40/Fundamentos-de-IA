```python
df = pd.DataFrame({
"Cidade": ["SP","SP","RJ","RJ"],
"Vendas": [10,20,30,40]
  })

df_agrupado = df.groupby("Cidade").mean()
display(df_agrupado)
```

```python
df = pd.DataFrame({
"Produto": ["Notebook","Mouse","Mouse","Teclado","Notebook"]
  })

df_agrupado = df.groupby('Produto').size()
display(df_agrupado)
```

```python
df = pd.DataFrame({
  "Produto": ["Notebook","Notebook","Mouse","Mouse"],
  "Cidade": ["SP","RJ","SP","RJ"],
  "Vendas": [10,15,20,25]
})

df_agrupado = df.groupby(["Produto", "Cidade"]).sum()
display(df_agrupado)
```

```python
df = pd.DataFrame({
"Produto": ["Notebook","Notebook","Mouse","Mouse"],
"Vendas": [10,50,20,25]
})

df_vendas_max = df.groupby("Produto").max()
display(df_vendas_max)
```

```python
df = pd.DataFrame({
"Produto": ["Notebook","Notebook","Mouse","Mouse"],
"Preco": [3000,3200,50,55]
})

df_medio = df.groupby("Produto").mean()
display(df_medio)
```

```python
df = pd.DataFrame({
"Cidade": ["SP","RJ","SP","BH","RJ"],
"Clientes": [100,150,200,80,120]
})

df_cidade = df.groupby("Cidade").sum()
display(df_cidade)
```

```python
df = pd.DataFrame({
"Departamento": ["TI","TI","RH","RH","Financeiro"],
"Salario": [5000,6000,4000,4200,7000]
})

display(df.groupby("Departamento").mean())
```

```python
df = pd.DataFrame({
"Categoria": ["Eletronico","Eletronico","Movel","Movel"],
"Quantidade": [5,10,3,7]
})

display(df.groupby("Categoria").sum())
```

```python
df = pd.DataFrame({
"Produto": ["Notebook","Notebook","Mouse","Mouse"],
"Ano": [2023,2024,2023,2024],
"Vendas": [10,15,20,25]
})

display(df.groupby(["Produto", "Ano"]).sum())
```

```python
df = pd.DataFrame({
"Aluno": ["Ana","Ana","Joao","Joao"],
"Nota": [7,9,6,8]
})

display(df.groupby("Aluno").mean())
```

```python
df = pd.DataFrame({
"Cidade": ["SP","SP","RJ","RJ"],
"Temperatura": [30,32,28,27]
})

display(df.groupby("Cidade").max())
```

```python
df = pd.DataFrame({
"Produto": ["Notebook","Notebook","Mouse","Mouse"],
"Quantidade": [5,2,10,8]
})

display(df.groupby("Produto").min())
```

```python
df = pd.DataFrame({
"Categoria": ["Eletronico","Eletronico","Movel","Movel"],
"Preco": [1000,1200,500,700],
"Quantidade": [2,3,1,4]
})

display(df.groupby(["Categoria", "Quantidade"]).agg(["mean", "sum"]))
```

```python
df = pd.DataFrame({
"Regiao": ["Norte","Sul","Norte","Sudeste","Sul"],
"Vendas": [100,200,150,300,250]
})

display(df.groupby("Regiao").sum().sort_values(by='Vendas' , ascending=False))
```
