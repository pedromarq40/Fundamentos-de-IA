```python
import numpy as np
import matplotlib.pyplot as plt
                            
# 1. Overfitting vs Underfitting
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.2, size=len(x))

# Fit simple (underfitting) model: linear
coef_linear = np.polyfit(x, y, 1)
y_linear = np.polyval(coef_linear, x)

# Fit complex (overfitting) model: high-degree polynomial
coef_poly = np.polyfit(x, y, 10)
y_poly = np.polyval(coef_poly, x)

plt.figure()
plt.scatter(x, y)
plt.plot(x, y_linear)
plt.plot(x, y_poly)
plt.title("Underfitting vs Overfitting")
plt.savefig('overfitting.png')
plt.close()

# 2. Regression comparison
x = np.linspace(0, 10, 100)
y = 3*x + 5 + np.random.normal(0, 5, size=len(x))

# Linear fit
coef_lin = np.polyfit(x, y, 1)
y_lin = np.polyval(coef_lin, x)

# Polynomial fit
coef_poly2 = np.polyfit(x, y, 3)
y_poly2 = np.polyval(coef_poly2, x)

plt.figure()
plt.scatter(x, y)
plt.plot(x, y_lin)
plt.plot(x, y_poly2)
plt.title("Linear vs Polynomial Regression")
plt.savefig('regressao.png')
plt.close()

# 3. ROC curve (simulated)
from sklearn.metrics import roc_curve
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

X, y = make_classification(n_samples=200, n_classes=2, random_state=0)
model = LogisticRegression()
model.fit(X, y)
y_scores = model.predict_proba(X)[:,1]

fpr, tpr, _ = roc_curve(y, y_scores)

plt.figure()
plt.plot(fpr, tpr)
plt.plot([0,1],[0,1])
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.savefig('roc_curve.png')
plt.close()
```
