```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris = load_iris()
# Carrega o dataset Iris (retorna um objeto do tipo Bunch, que funciona como um dicionário)

print(type(iris))
```

```python
iris = load_iris()

print(iris.keys())

# Acessa a descrição do dataset ('DESCR') e imprime apenas os primeiros 193 caracteres.
# O '\n...' adiciona uma quebra de linha e reticências para indicar que o texto foi cortado.
print(iris['DESCR'][:193] + '\n...')
```

```python
iris_datasset = load_iris()
iris_X = iris_datasset.data
iris_Y = iris_datasset.target

np.random.seed(0)
indices = np.random.permutation(len(iris_X))

iris_X_train = iris_X[indices[:-10]]
iris_Y_train = iris_Y[indices[:-10]]
iris_X_test = iris_X[indices[-10:]]
iris_Y_test = iris_Y[indices[-10:]]

knn = KNeighborsClassifier()
knn.fit(iris_X_train, iris_Y_train)
KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minwoski',
                     metric_params=None, n_jobs=1, n_neighbors=5, p=2,
                     weights='uniform')

predictedclasses = knn.predict(iris_X_test)

classifier_accuracy = np.mean(predictedclasses == iris_Y_test)
print(classifier_accuracy)

```
