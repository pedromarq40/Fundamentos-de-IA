```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target,
random_state=100)

knn = KNeighborsClassifier(n_neighbors=5, weights='distance', algorithm='auto',
    leaf_size=30, metric='minkowski', p=2, n_jobs=1)

knn.fit(x_train, y_train)
test = knn.predict(x_test)
accuracy = np.mean(test == y_test)
print(f"Accuracy: {accuracy}")
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import svm

iris = load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=100)

C = 1.0
svc = svm.SVC(kernel='linear', C=C)
svc.fit(x_train, y_train)
test = svc.predict(x_test)
accuracy = np.mean(test == y_test)
print(f"Accuracy: {accuracy:.2%}")
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(knn, iris.data, iris.target, cv=10)
print(scores)
print(f"CV Média: {scores.mean():.2%} ± {scores.std():.2%}")

scores2 = cross_val_score(svc, iris.data, iris.target, cv=15)
print(scores2)
print(f"CV Média: {scores2.mean():.2%} ± {scores2.std():.2%}")
```
