### Avaliar a qualidade de dados do dataset
    Passos:
        - 1. Copiar Dataset
        - 2. Substituir valors 0 por NaN 
            * Para este caso em especial, em outras bases de dados pode ser necesário outro tratamento). Porque não faz sentido biologicamente, exemplos:
                - Glicose = 0 => Impossível
                - Pressão sanguínea = 0 => Impossível

Isso é uma etapa clássica da limpeza dos dados, em várias projetos de Ciência de Dados fazemos essa limpeza antes de ir para a fase de pré-processamento (a ser explicaa a posteriori).

````python
import numpy as np
import pandas as pd

file_path = 'diabetes.csv'
df = pd.read_csv(file_path)
print(df)

#Copia a base de dados
df_clean_copy = df.copy(deep = True)
df_clean_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df_clean_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
````

## GroupBy no Pandas

O **groupby** do pandas é usado para **agrupar linhas de um DataFrame com base em uma ou mais colunas e aplicar operações nesses grupos** (soma, média, contagem, etc.).

Ele segue o conceito clássico chamado:

**Split → Apply → Combine**

1. **Split** – divide os dados em grupos  
2. **Apply** – aplica uma função em cada grupo  
3. **Combine** – junta os resultados  

---

O groupby serve para:

- agrupar dados
- aplicar operações estatísticas
- gerar relatórios agregados

É muito utilizado em:
 - análise de dados
 - ciência de dados
 - finanças
 - relatórios de vendas

# 1. Exemplo simples de groupby

```python
import pandas as pd

df = pd.DataFrame({
    "produto": ["A", "B", "A", "B", "C"],
    "vendas": [10, 5, 3, 2, 8]
})

resultado = df.groupby("produto").sum()

print(resultado)

````python
import pandas as pd

df = pd.DataFrame({
    "produto": ["A", "B", "A", "B", "C"],
    "vendas": [10, 5, 3, 2, 8]
})

resultado = df.groupby("produto").sum()

print(resultado)
````

### Outras operações comuns com group by

````python
import pandas as pd

df = pd.DataFrame({
    "produto": ["A", "B", "A", "B", "C"],
    "vendas": [10, 5, 3, 2, 8]
})

resultado = df.groupby("produto").sum()

print(resultado)
````

````python
#Soma
df.groupby("produto")["vendas"].sum()
````

````python
#Média
df.groupby("produto")["vendas"].mean()
````

````python
#Mínimo
df.groupby("produto")["vendas"].min()
````

````python
#Máximo
df.groupby("produto")["vendas"].max()
````

````python
#Quantidade de Itens
df.groupby("produto")["vendas"].count()
````

````python
## Agrupar por mais de uma coluna
df = pd.DataFrame({
    "produto": ["A","A","B","B"],
    "ano": [2023,2024,2023,2024],
    "vendas": [10,12,5,8]
})

df.groupby(["produto","ano"]).sum()
````

````python
## Usando várias funções ao mesmo tempo
df.groupby("produto")["vendas"].agg(["sum","mean","max"])
````

````python
#Criando colunas agregadas personalizadas
df.groupby("produto").agg(
    total_vendas=("vendas","sum"),
    media_vendas=("vendas","mean")
)
````

````python
# Recriando o index
df.groupby("produto")["vendas"].sum().reset_index()
````

## Pandas com Base de Dados

Utilizaremos neste exemplo a criação de uma base de dados em SQLite que pode ser aberto com o DB Browser (https://sqlitebrowser.org/). Sugiro baixá-lo para treinarmos a navegação dos dados usando a linguagem SQL, que é padrão de mercado e largamento utilizada na indústria.

Neste notebook, iremos fazer as chamadas SQL usando a api, mas vocês podem refletir os mesmos comandos no DB Browser.

````python
import pandas as pd
import sqlite3

df = pd.read_csv("diabetes.csv")

#Criando a base de dados
conn = sqlite3.connect("diabetes.db")

#Fazendo o DUMP dos dados do arquivo csv na base de dados
df.to_sql("patients", conn, if_exists="replace", index=False)

#"patients" → nome da tabela
# replace → substitui se existir
# index=False → não salva o índice
````

Agora abra os arquivos no DB Browser

````python
import pandas as pd
import sqlite3

conn = sqlite3.connect("diabetes.db")

query = """
SELECT *
FROM patients
LIMIT 10
"""

pd.read_sql(query, conn)
````

````python
query = """
SELECT age 
FROM patients 
"""

pd.read_sql(query, conn)
````

````python
query = """
SELECT Glucose, BMI, Age
FROM patients
LIMIT 10
"""

pd.read_sql(query, conn)
````

````python
# CLÁUSULA WHERE
query = """
SELECT *
FROM patients
WHERE Glucose > 150
"""
pd.read_sql(query, conn)
````

````python
query = """
SELECT Glucose, BMI, Age
FROM patients
WHERE Glucose > 150
AND BMI > 30
"""

pd.read_sql(query, conn)
````

````python
# ORDENANDO OS DADOS
## ORDEM DESCRESCENTE
query = """
SELECT *
FROM patients
ORDER BY Glucose DESC
LIMIT 10
"""

pd.read_sql(query, conn)
````

````python
# ORDENANDO OS DADOS
## ORDEM CRESCENTE
query = """
SELECT *
FROM patients
ORDER BY BloodPressure
LIMIT 50
"""

pd.read_sql(query, conn)
````

````python
# Agrupando os Dados
query = """
SELECT Outcome, AVG(Glucose) as media_glucose
FROM patients
GROUP BY Outcome
"""

pd.read_sql(query, conn)
````

````python
# Contando os Pacientes
query = """
SELECT Outcome, COUNT(*) as total
FROM patients
GROUP BY Outcome
"""

pd.read_sql(query, conn)
````

````python
# Criando uma nova tabela - Com pacientes com alto risco

query = """
CREATE TABLE high_risk AS
SELECT *
FROM patients
WHERE Glucose > 160
AND BMI > 35
"""

conn.execute(query)
````

````python
query = """
SELECT *
FROM high_risk
"""

pd.read_sql(query, conn)
````

````python
# Fechando a conexão
conn.close()
````
