# **Máquina de Vetor de Suporte - SVM**

Máquina de Vetor de Suporte (SVM – Support Vector Machine) é um algoritmo de aprendizado de máquina muito usado para classificação e também para regressão. A ideia principal da SVM é encontrar a melhor forma de separar os dados em diferentes categorias.

Imagine que você tem pontos em um gráfico, representando duas classes (por exemplo, “doente” e “saudável”). A SVM tenta desenhar uma linha (ou, em dimensões maiores, um hiperplano) que separa esses pontos da melhor maneira possível.

O diferencial da SVM é que ela não escolhe qualquer linha separadora: ela escolhe aquela que maximiza a margem, ou seja, a distância entre a linha de separação e os pontos mais próximos de cada classe. Esses pontos mais próximos são chamados de vetores de suporte, pois são eles que “definem” a posição da fronteira de decisão.

Em muitos casos reais, os dados não são separáveis por uma linha reta. Para resolver isso, a SVM utiliza uma técnica chamada truque do kernel. Essa técnica transforma os dados para um espaço de maior dimensão, onde eles podem ser separados linearmente. Alguns kernels comuns são:

Linear
Polinomial
RBF (Radial Basis Function)

Além disso, a SVM possui um parâmetro importante chamado C, que controla o equilíbrio entre:

Maximizar a margem
Permitir alguns erros de classificação

Um valor alto de C tenta classificar todos os pontos corretamente (mesmo que a margem fique menor), enquanto um valor baixo permite mais erros, mas com uma margem maior.



### Resumo, a SVM é um algoritmo poderoso porque:

- Funciona bem em espaços de alta dimensão
- É eficaz mesmo com poucos dados
- Pode lidar com problemas não lineares usando kernels

Por isso, ela é muito utilizada em áreas como reconhecimento de imagens, bioinformática e análise de texto.

## 🧠 1. Introdução Teórica

A **Máquina de Vetor de Suporte (SVM)** é um algoritmo de aprendizado supervisionado usado para:
- Classificação
- Regressão

### 🎯 Ideia principal
Encontrar um **hiperplano** que separa classes maximizando a **margem**.

- Margem = distância entre a fronteira e os pontos mais próximos
- Esses pontos são os **vetores de suporte**

---

## 📐 2. Conceitos Fundamentais

### 🔹 Hiperplano
- Em 2D: uma reta
- Em 3D: um plano
- Em dimensões maiores: hiperplano

### 🔹 Vetores de Suporte
São os pontos mais próximos da fronteira.
Eles determinam completamente o modelo.

### 🔹 Parâmetro C
Controla o trade-off:
- C alto → menos erros, menor margem (overfitting)
- C baixo → mais margem, mais tolerância a erros

### 🔹 Kernel Trick
Permite separar dados não lineares:
- linear
- rbf
- polynomial

````python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import svm
import numpy as np

# Carregar dados
iris = load_iris()

# Divisão treino/teste
X_train, X_test, y_train, y_test = train_test_split(
    iris['data'], iris['target'], random_state=0
)

# Definir parâmetro C
C = 1.0

# SVC com kernel linear
svc = svm.SVC(kernel='linear', C=C).fit(X_train, y_train)

# LinearSVC (kernel linear otimizado)
lin_svc = svm.LinearSVC(C=C, max_iter=10000).fit(X_train, y_train)

# SVC com kernel RBF
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X_train, y_train)

# SVC com kernel polinomial (grau 3)
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X_train, y_train)

# Predição
y_pred = svc.predict(X_test)

# Resultado
print("Predições:", y_pred)

# Acurácia
classifier_score = np.mean(y_pred == y_test)
print("Acurácia:", classifier_score)
````

````python
#Avaliando todos os modelos
models = {
    "SVC Linear": svc,
    "LinearSVC": lin_svc,
    "RBF": rbf_svc,
    "Polinomial": poly_svc
}

for name, model in models.items():
    score = model.score(X_test, y_test)
    print(f"{name}: {score:.2f}")
````

## Visualizando a Fronteira de Decisão
Visualizar a fronteira de decisão é uma das formas mais intuitivas de entender como a SVM funciona 👇

Como o dataset Iris tem 4 dimensões, precisamos reduzir para 2 features para conseguir plotar. Vamos usar só as duas primeiras.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Carregar dados
iris = load_iris()

# Usar apenas 2 features (para visualização)
X = iris.data[:, :2]
y = iris.target

# Dividir treino/teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=0
)

# Treinar SVM
model = svm.SVC(kernel='linear', C=1.0)
model.fit(X_train, y_train)

# Criar grade de pontos
h = 0.02  # passo
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)

# Prever para cada ponto da grade
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotar
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)

# Plotar pontos reais
plt.scatter(X[:, 0], X[:, 1], c=y)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Fronteira de decisão - SVM")

plt.show()
````

O que você vai ver
O fundo colorido → regiões de decisão da SVM
Os pontos → seus dados reais
A “divisão” → onde a SVM separa as classes

### Criando um modelo mais completo que possui:

- Visualização da fronteira de decisão
- Destaque dos vetores de suporte (em vermelho)
- Uso com 2 features (visual direto)
- Uso com PCA (mais realista)
- Comparação com kernel RBF

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

# 2. Carregar dados
iris = load_iris()
X = iris.data
y = iris.target

# =============================================
# PARTE 1 - VISUALIZAÇÃO COM 2 FEATURES
# =============================================

# Selecionar apenas 2 features
X2 = X[:, :2]

# Divisão treino/teste
X_train, X_test, y_train, y_test = train_test_split(
    X2, y, random_state=0
)

# Treinar modelo SVM
model = svm.SVC(kernel='linear', C=1.0)
model.fit(X_train, y_train)

# Criar grade
h = 0.02
x_min, x_max = X2[:, 0].min() - 1, X2[:, 0].max() + 1
y_min, y_max = X2[:, 1].min() - 1, X2[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)

# Predição na grade
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)

# Pontos reais
plt.scatter(X2[:, 0], X2[:, 1], c=y, edgecolors='k')

# DESTACAR VETORES DE SUPORTE
support_vectors = model.support_vectors_
plt.scatter(
    support_vectors[:, 0],
    support_vectors[:, 1],
    s=150,
    facecolors='none',
    edgecolors='red',
    linewidths=2,
    label='Vetores de Suporte'
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVM com Vetores de Suporte destacados")
plt.legend()
plt.show()

# Acurácia
print("Acurácia (2 features):", model.score(X_test, y_test))


# =============================================
# PARTE 2 - USANDO PCA (TODAS AS FEATURES)
# =============================================

# Reduzir para 2D com PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, random_state=0
)

# Treinar SVM com kernel RBF
model_pca = svm.SVC(kernel='rbf', gamma=0.7, C=1.0)
model_pca.fit(X_train, y_train)

# Criar grade
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)

# Predição
Z = model_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)
print("Número de vetores de suporte (modelo PCA):", len(model_pca.support_vectors_))
````
