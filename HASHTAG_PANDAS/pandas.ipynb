<h3>Fundamentos de Pandas</h3>

```python
import pandas as pd
import numpy as np

dados = {
    'Produto': ['Celular', 'Notebook', 'Tablet', 'Fone'],
    'Preco': [2000, 5000, 1500, 300],
    'Quantidade': [10, 5, 20, 50]
}

df = pd.DataFrame(dados)

print(df)
print('\n')

print(df['Preco'])
```

```python
#Criando faturamento

df['Faturamento'] = df['Preco'] * df['Quantidade']

print(df)
print('\n')

print(df.describe())
```

<h3>Agrupamento</3>

```python
import pandas as pd

# 1. Dados mais complexos (Várias vendas do mesmo item)
dados = {
    'Categoria': ['Eletrônicos', 'Eletrônicos', 'Móveis', 'Móveis', 'Eletrônicos', 'Móveis', 'Móveis'],
    'Produto': ['Celular', 'Notebook', 'Mesa', 'Cadeira', 'Celular', 'Mesa', 'Cadeira'],
    'Vendas': [2000, 5000, 500, 200, 2200, 400, 500]
}
df = pd.DataFrame(dados)

print("--- Tabela Original ---")
print(df)
print("\n")

# 2. A MÁGICA DO GROUPBY
# Agrupe por 'Categoria' e some as 'Vendas'
df_agrupado = df.groupby('Categoria')

resumo = df_agrupado['Vendas'].sum()

print("--- Resumo por Categoria ---")
print(resumo)
print('\n')

# Agrupe por 'Produto' e somea as vendas

venda_por_produto = df.groupby('Produto')['Vendas'].agg(['sum', 'count'])

print(venda_por_produto)
```

<h3>Agregate</h3>

```python
relatorio_completo = df.groupby('Categoria')['Vendas'].agg(['sum', 'mean', 'count'])

print(relatorio_completo)
```

```python
relatorio_completo = df.groupby('Produto')['Vendas'].agg(['sum', 'mean', 'max', 'min', 'count', 'idxmax'])

print(relatorio_completo)
```

```python
dados = {
    'Categoria': ['Eletrônicos', 'Eletrônicos', 'Móveis', 'Móveis', 'Eletrônicos'],
    'Produto': ['Celular', 'Notebook', 'Mesa', 'Cadeira', 'Celular'],
    'Vendas': [2000, 5000, 500, 200, 2200],
    'Quantidade Por Venda': [2, 1, 4, 10, 2]
}
df = pd.DataFrame(dados)

print(df)
print('\n')

grupo_produto = df.groupby('Produto')['Vendas'].agg(['sum', 'count'])

print(grupo_produto)
print('\n')

relatorio = df.groupby('Categoria').agg({
    'Vendas': ['sum', 'count', 'mean'],   
    'Quantidade Por Venda': ['sum', 'count', 'mean'] 
})

print(relatorio)
```

```python
desafio = df.groupby('Produto').agg({
    'Vendas': 'min', 'Quantidade Por Venda' : 'sum'
})

print(desafio)
```

<h3>Ordem crescente e decrescente / sort_values() </h3>

```python
ranking = venda_por_produto.sort_values(by='',ascending=False)

print(ranking)
```

```python
ranking = venda_por_produto.sort_values(ascending=True)

print(ranking)
```

<h3>Merge</h3>

```python
vendas = pd.DataFrame({
    'Data': ['01/01', '02/01', '03/01', '03/01'],
    'Produto': ['Celular', 'Notebook', 'Celular', 'Fone'],
    'Valor': [2000, 5000, 3000, 300],
    'id_cliente': [1, 2, 1, 3]  
})

clientes = pd.DataFrame({
    'id_cliente': [1, 2, 3, 4],
    'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
    'Email': ['joao@email.com', 'maria@email.com', 'pedro@email.com', 'ana@email.com']
})

print("--- Vendas ---")
print(vendas)
print("\n--- Clientes ---")
print(clientes)
```

```python
vendas_completas = pd.merge(vendas, clientes, on='id_cliente', how='left')

print(vendas_completas)
print('\n')

vendas_completas = pd.merge(clientes, vendas, on='id_cliente', how='left')

print(vendas_completas)
```
