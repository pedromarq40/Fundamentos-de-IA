Métodos gerais

* df.groupby() - Agrupar series por atributo
* df.agg() - Aplicar vários métodos, vem com o groupby
* df.sort_values(by='Colunas/Parametros') - Organizar Valores, recebe uma lista com as colunas usadas como comparação
* pd.merge() - Mesclar tabelas em alguma coluna igual
* pd.read() - Ler csv ou excel
* df.describe() - Estatísticas da tabela
* df.info() - Tipos de dados e valores
* df.dropna() - Tirar todas as linhas com NaN
* df.drop() - Tirar linha ou coluna
* df.fillna() - Trocar NaN por algum valor
* pd.to_datetime( , format=) - Transformar data str para um objeto próprio
* df[coluna].str.metodo() - Formatar Coluna
* df.drop_duplicates(keep=, subset=)
* np.where(df[coluna] + condicao, true, false) - Aplicar condição a uma coluna
* .unique() - Ver todos os valores diferentes
* df.loc(index, coluna) - Localizar informação especifica com nome/index(pode variar caso reorganizado) e coluna
* df.iloc(index, index-coluna) - Localizar informação especifica com index e index da coluna
* df.reset_index(drop=True) - Reorganiza index e joga os antigos fora
* df.to_*formato*(index=bool) - Salvar df com ou sem index
* df.pivot_table(index='Categoria', columns='Produto', values='Vendas', aggfunc='sum', fill_value=0)
* df['column'],dt.to_period('M'/'Y'/'D')
