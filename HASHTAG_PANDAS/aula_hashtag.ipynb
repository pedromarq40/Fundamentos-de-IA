<h3>Leitura de Arquivos</h3>

```python
import numpy as np
import pandas as pd

df_vendas = pd.read_csv('vendas_tech.csv', low_memory=False)
df_gerentes = pd.read_excel('gerentes_lojas.xlsx')

# Configura para mostrar no MÁXIMO 20 linhas
pd.set_option('display.max_rows', 15)

# Configura para mostrar no MÁXIMO 50 colunas (evita barra de rolagem infinita lateral)
pd.set_option('display.max_columns', 50)

display(df_vendas.head(15))
display(df_gerentes)
```

anotações:

parametros do read:
encoding= (utf-8, latin1,), para decodificar acentos etc
sep = ( ; , ), Separação das informações csv
low_memory = ( true, false), consumir memória para separar os tipos de dados str floatetc

```python
#ferramentas de inspeção

display(df_vendas)
#display(df_vendas.head(5)) ver as n primeiras linhas
#display(df_vendas.tail(5)) ver as n primeiras linhas
# display(df_vendas.sample(5)) ver 5 aleatórias
display(df_vendas.shape) # linhas x colunas
display(list(df_vendas.columns)) # lista das colunas
```

```python
display(df_vendas.info()) # info gerais
print('\n')
print('\n')
display(df_vendas.describe()) # descrição detalhada
```

<h3>Tratamento de dados</h3>

```python
display(df_vendas['Loja'])

display(df_vendas[['Loja', 'Cliente']]) # Deve ser uma LISTA DE COLUNAS
```

<h3>Tratamento de Colunas</h3>

```python
df_vendas_analise = df_vendas.drop(columns=['Data_Base']) # Excluir linhas ou colunas

display(df_vendas_analise)
```

para retirar posso ou passar as listas de colunas desejadas como indice ou
df.drop(), o parametro deve ser uma lista com as colunas a retirar


<h3>Eliminar valores Nulos</h3>

```python
display(df_vendas_analise.info()) #Ver colunas com NaN
display('\n')

#DELETAR LINHAS COM QUALQUER NAN
df_vendas_analise = df_vendas_analise.dropna() 

display(df_vendas_analise)
display('\n')

# TRATAR APENAS OS VALORES NAN DA COLUNA LOJA
df_vendas_analise = df_vendas.drop(columns=['Data_Base'])
df_vendas_analise['Loja'] = df_vendas['Loja'].fillna('Online') # Coloca o valor do parametro no lugar do nan

display(df_vendas_analise)
```

<h3>Tipos de Dados Diferentes</h3>

```python
#Transformando a coluna de data em datas reais

df_vendas_analise['Data'] = pd.to_datetime(df_vendas_analise['Data'], format='%Y-%m-%d')

display(df_vendas_analise)
display(df_vendas_analise.info())
```

<h3>Padronização / Formatação</h3>

```python
df_vendas_analise['Loja'] = df_vendas_analise['Loja'].str.strip()
df_vendas_analise['Loja'] = df_vendas_analise['Loja'].str.title()

display(df_vendas_analise.head(20))
```

<h3>Remover Duplicatas</h3>

```python
df_vendas_analise = df_vendas_analise.drop_duplicates(subset=['ID_Pedido', 'Loja'], keep='last')
display(df_vendas_analise)
```

subset : colunas para considerar iguais
keep : first ou last, remover primeira ou segunda duplicata

<h3>Criar Novas Colunas</h3>

```python
df_vendas_analise['Faturamento'] = df_vendas_analise['Qtd'] * df_vendas_analise['Preco_Unitario']
display(df_vendas_analise)
```

<h3>Condicionais / Tratamento</h3>

```python
df_vendas_loja = df_vendas_analise.groupby('Loja').agg({'Cliente':'count', 'Faturamento':'sum'})

print(df_vendas_loja)
```

```python
df_vendas_analise['Forma_De_Venda'] = np.where(df_vendas_analise['Loja'] == 'Online', 'Online', 'Presencial')
display(df_vendas_analise)
```

```python
display(df_vendas_analise['Loja'].unique())

sudeste = ['São Paulo', 'Belo Horizonte', 'Rio de Janeiro']
nordeste = ['Salvador', 'Recife']
sul = ['Curitiba', 'Porto Alegre']

dict_regioes = {
    'São Paulo': 'Sudeste',
    'Belo Horizonte': 'Sudeste',
    'Online' : 'Online',
    'Rio De Janeiro': 'Sudeste',
    'Salvador' : 'Nordeste',
    'Recife' : 'Nordeste',
    'Curitiba' : 'Sul',
    'Porto Alegre' : 'Sul',
}

#df_vendas_analise['Região'] = np.where(df_vendas_analise['Loja'] in sudeste, 'Sudeste', np.where(...)) JEITO 1
df_vendas_analise['Região'] = df_vendas_analise['Loja'].map(dict_regioes)

display(df_vendas_analise)
```

<h3>Exportar Pedaços da Tabela / Filtrar</h3>

```python
df_vendas_analise = df_vendas_analise.sort_values(by=['Data', 'Faturamento'], ascending=True)

loja = df_vendas_analise.loc[4, 'Loja']
produto = df_vendas_analise.loc[4, 'Produto']
cliente = df_vendas_analise.loc[4, 'Cliente']
print(loja, produto, cliente)
```

```python
loja = df_vendas_analise.iloc[4, 2]
produto = df_vendas_analise.iloc[4, 3]
cliente = df_vendas_analise.iloc[4, 6]
print(loja, produto, cliente)

display(df_vendas_analise)
```

```python
df_vendas_analise = df_vendas_analise.reset_index(drop=True)
display(df_vendas_analise)
```

* df.loc - Localizar informação especifica com index e coluna
* df.iloc(index, index-coluna) - Localizar informação especifica com index e index da coluna

<h3>Análise com Agrupamento</h3>

```python
#df_analise_loja = df_vendas_analise.groupby('Loja')['Faturamento'].sum()
df_analise_loja = df_vendas_analise[['Loja', 'Faturamento']].groupby('Loja').sum().sort_values(by='Faturamento', ascending=False)
df_analise_loja = df_analise_loja.reset_index()
df_analise_loja['Faturamento'] = df_analise_loja['Faturamento'].map(lambda x : f'R${x:,.2f}')
display(df_analise_loja)
```

```python
df_vendas_online = df_vendas_analise[df_vendas_analise['Loja'] == 'Online']
df_vendas_online = df_vendas_online[['Produto', 'Qtd']].groupby('Produto').sum().sort_values(by='Qtd', ascending=False)

df_vendas_online = df_vendas_online.rename(columns={'Qtd':'Vendas Totais'})

display(df_vendas_online)
```

```python
df_vendas_por_loja = df_vendas_analise[['Loja', 'Produto', 'Qtd']].groupby(['Loja', 'Produto']).sum()

with pd.option_context('display.max_rows', None):
    display(df_vendas_por_loja)
```

```python
df_nsei = df_vendas_analise[['Loja', 'Produto', 'Qtd']].groupby(['Produto', 'Loja']).sum().sort_values(by=['Produto', 'Qtd'], ascending=[True, False])

df_nsei.to_csv('produto_por_regiao.csv')

with pd.option_context('display.max_rows', None):
    display(df_nsei)
```

```python
#display(df_vendas_analise.head())

# quais gerentes bateram a meta em janeiro de 23

#display(df_gerentes)

df_meta = df_vendas_analise[(df_vendas_analise['Data'].dt.year == 2023) & (df_vendas_analise['Data'].dt.month == 1)]

df_meta = df_meta[['Faturamento','Loja']].groupby(['Loja']).sum()

df_meta = df_meta.reset_index(drop=False)

df_meta = df_meta.sort_values(by=['Loja'], ascending=True)

df_meta_gerentes = pd.merge(df_meta, df_gerentes, how='left', on='Loja') #esse e o de baixo fazem o mesmo

display(df_meta_gerentes)

df_meta_gerentes = df_meta.merge(df_gerentes, on='Loja', how='left') #esse e o de baixo fazem o mesmo

display(df_meta_gerentes)

# NAO ENCONTRA RJ PELA DIFERENÇA DE FORMATAÇÂO

df_gerentes['Loja'] = df_gerentes['Loja'].str.strip().str.title()

display(df_gerentes)

df_meta_gerentes = df_meta.merge(df_gerentes, on='Loja', how='left') 

df_meta_gerentes['Bateu_Meta'] = np.where(df_meta_gerentes['Faturamento'] >= df_meta_gerentes['Meta_Mensal'], 'Sim', 'Não')

display(df_meta_gerentes)
```
