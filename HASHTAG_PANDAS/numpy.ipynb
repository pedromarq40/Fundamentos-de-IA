<h3>Fundamentos de Numpy</h3>

```python
import numpy as np

vetor = np.array([1, 2, 3, 4])

print(vetor)
```

```python
precos = np.array([10, 20, 30])
quantidades = np.array([3, 4, 5])

print(precos * quantidades)
```

```python
sequencia = np.arange(10)
print(f'Sequencia: {sequencia}')


dados_fake = np.random.randint(0, 100, size=5)
print(f'Dados fake: {dados_fake}')


print(f'Média: {dados_fake.mean()}')
print(f'Maior: {dados_fake.max()}')
print(f'Menor: {dados_fake.min()}')
```

```python
#Filtração booleana

notas = np.array([2, 4, 6, 7, 8, 2, 9, 7])

print(notas >= 6.0)
filtro = notas >= 6.0

aprovados = notas[filtro]
print(f'Aprovados: {aprovados}')
```

<h3>Matrizes</h3>

```python
notas = np.array([
    [8.0, 7.5, 9.5],
    [3.0, 10.0, 7.0]
])

print(notas)
print(f'Linhas e colunas: {notas.shape}')
```

```python
nota_especifica = notas[1, 2]
print(nota_especifica) 


aluno1 = notas[0, :]
print(aluno1)
```

<h3>mean(axis=0): Média da coluna
mean(axis=1): Média da linha</h3>

```python
medias_alunos = notas.mean(axis=1)
print(medias_alunos)

media_vertical = notas.mean(axis=0)
print(media_vertical)
```

```python
matriz = np.array([[10, 20], [30, 40]])
print(matriz[1,0])
print(matriz.sum()) #Soma de todos
print(matriz.sum(axis=1)) #Soma da linha
```
